```python
class Node:
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
        
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertNext(self,node):
        if self.head is None:
            self.head = node
        else:
            curr_node = self.head 
            while curr_node.next is not None:
                curr_node = curr_node.next
            
            curr_node.next = node
            
    
    def insertAfter(self,afterNode,newNode):
        originalNode = afterNode.next
        afterNode.next = newNode
        newNode.next = originalNode
        
    def insertHead(self,newNode):
        if self.head is None:
            self.head = newNode
            return
        tempNode = self.head
        self.head = newNode
        newNode.next = tempNode
        
        
    def insertAt(self,pos,newNode):
        counter = 1
        curr_node = self.head
        while counter < pos-1:
            curr_node = curr_node.next
            counter = counter+1
            if curr_node.next is None:
                print("The index does not exist so appending node in the end")
                curr_node.next = newNode
                return
                
        tempNode = curr_node.next
        curr_node.next = newNode
        newNode.next = tempNode
    
    
    def deleteNode(self, node):
        curr_node = self.head
        while True:
            if curr_node.next == node:
                target_node = curr_node.next
                curr_node.next = target_node.next
                return
            elif curr_node.next is None:
                print("The node you want to delete isnt in the list")
                return
            else:
                curr_node = curr_node.next
                
        
    def printlist(self):
        if self.head is None:
            print("Linked List is empty")
            return
        currentNode = self.head
        while True:
            if currentNode is None:
                break
            print("-------------------------------------------------------------------------------------------")
            print(f'|   {currentNode.data} ({currentNode})  |   {currentNode.next}   |')
            print("-------------------------------------------------------------------------------------------")
            
            currentNode = currentNode.next
    

    

newNode = Node("soham")
linked = LinkedList()
linked.insertNext(newNode)

secondNode = Node("aman")
linked.insertNext(secondNode)

thirdNode = Node("prachi")
linked.insertNext(thirdNode)

fourthNode = Node("tasmai")
linked.insertAfter(secondNode,fourthNode)

headNode = Node("teri mummy")
linked.insertHead(headNode)

randomNode = Node("test")
linked.insertAt(5,randomNode)

newNode = Node("bruh")
linked.deleteNode(newNode)


linked.printlist()
```
