# Queue
A queue is basically an abstraction and a constraint version of a linkedlist. It follows the First in and first out protocol. Just like a queue in real life, a new item is added to the tail and an item is removed from the heaad. 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.index = 0
        
    def enqueue(self, newNode):
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode
        self.index += 1
            
    def deqeue(self):
        self.head = self.head.next
        self.index -= 1
        
    def showQueue(self):
        currentNode = self.head
        while True:
            print(f'{currentNode.data} >>>> {currentNode.next}')
            currentNode = currentNode.next
            if currentNode is None:
                break
        
        print('\n')
        
            
    def printHeadTail(self):
        print(f"HEAD ----> {self.head.data}")
        print(f"TAIL ----> {self.tail.data}")
        print(self.index)
        print('\n')
    
            

newNode = Node("Soham")
secondNode = Node('Shri')
thirdNode = Node("Iqra")
fourthnode = Node("Sanjana")


queue = Queue()
queue.enqueue(newNode)
queue.enqueue(secondNode)
queue.enqueue(thirdNode)
queue.enqueue(fourthnode)

queue.showQueue()
queue.printHeadTail()

queue.deqeue()
queue.printHeadTail()
```
